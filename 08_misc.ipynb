# RESHAPING AND PIVOTING

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("sales_by_employee.csv", parse_dates=["Date"])
sales
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="count"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values="Revenue",
    aggfunc="count",
    fill_value=0,
    margins=True,
    margins_name = "Total"
)
```

```python
sales_by_name_date = sales.pivot_table(
    index="Name",
    columns="Date",
    values="Revenue",
    aggfunc="sum"
)
sales_by_name_date
```

```python
sales_by_name_date.stack()
```

```python
sales_by_name_company = sales.pivot_table(
    index=["Name", "Customer"],
    values="Revenue",
    aggfunc="sum"
)
sales_by_name_company
```

```python
sales_by_name_company.unstack()
```

# MERGING, JOINING AND CONCATENATION

```python
groups1 = pd.read_csv("groups1.csv")
groups2 = pd.read_csv("groups2.csv")
categories = pd.read_csv("categories.csv")
cities = pd.read_csv("cities.csv", dtype={"zip":"string"})
```

```python
groups1.head(1)
```

```python
groups2.head(1)
```

```python
groups = pd.concat(objs=[groups1, groups2], ignore_index=True)
```

```python
groups.head(1)
```

```python
categories.head(1)
```

## JOINS

```python
groups.merge(categories, how="inner", on="category_id", indicator=True)
```

```python
groups.head(1)
```

```python
cities.head(1)
```

```python
groups.merge(cities, how="outer", left_on="city_id", right_on="id", indicator=True)
```

```python
cities = cities.set_index("id")
```

```python
groups.merge(cities, how="left", left_on="city_id", right_index=True)
```

# WORKING WITH JSON FILES

```python
nobel = pd.read_json("nobel.json")
nobel
```

```python
type(nobel.loc[2, "prizes"])
```

```python
chemistry_2019 = nobel.loc[0, "prizes"]
type(chemistry_2019)
chemistry_2019
```

```python
pd.json_normalize(data=chemistry_2019)
```

```python
pd.json_normalize(data=chemistry_2019, record_path="laureates")
```

```python
pd.json_normalize(
    data=chemistry_2019,
    record_path="laureates",
    meta=["year", "category"]
)
```

```python
def add_laureates_key(entry):
    entry.setdefault("laureates", [])

nobel["prizes"].apply(add_laureates_key)
```

```python
winners = pd.json_normalize(
    data=nobel["prizes"],
    record_path="laureates",
    meta=["year", "category"]
)
winners
```

```python
type(winners)
```

```python
winners.to_json("winners.json",orient="records")
```
